```python
#I'll define the threshold values
low_threshold=0.3
medium_threshold=0.5
# create the new categorical column called DTI_category
df['DTI_category'] = pd.cut(df['DTIRatio'], bins=[0, low_threshold, medium_threshold, float('inf')],
                            labels=['low', 'medium', 'high'])

print(df['DTI_category'].head())
```

```python
df.head() # great! created a new column dti_category to categorize the dtiratio as medium, high or low.
```

```python
#Define credit_score ranges

# Define credit score ranges
credit_score_bins = [300, 500, 600, 700, 800, 900]

# Create bins for credit scores and calculate average loan amount for each bin
df['credit_score_range'] = pd.cut(df['CreditScore'], bins=credit_score_bins)
avg_loan_by_credit_score = df.groupby('credit_score_range')['LoanAmount'].mean()
avg_loan_by_credit_score
```
